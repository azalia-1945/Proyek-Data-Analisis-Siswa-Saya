Apakah benar performa laki-laki dan perempuan ujian itu berbeda dengan satu sama lain? Mari kita buktikan

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
data = pd.read_csv('/content/StudentsPerformance  (1).csv')
```

```python
data.info()
```

1. Perbandingan nilai matematika, membaca, dan menulis laki-laki dan perempuan

a. Nilai matematika

```python
data.groupby('gender')['math score'].agg(['min', 'max', 'mean', 'count'])
```

Dilihat dari data berikut, perempuan memiliki nilai minimum 0 dan maksimum 100, dengan rata-rata nilai dari 518 anggota yaitu 63.6. Sedangkan laki-laki memiliki nilai minimum 27 dan maksimum 100, dengan rata-rata nilai dari 482 anggota yaitu 68.7. Berikut visualisasi datanya:

```python
sns.histplot(data, x='math score', hue='gender')
```

b. Nilai membaca

```python
data.groupby('gender')['reading score'].agg(['min', 'max', 'mean', 'count'])
```

Dilihat dari data berikut, perempuan memiliki nilai minimum 17 dan maksimum 100, dengan rata-rata nilai dari 518 anggota yaitu 72.6. Sedangkan laki-laki memiliki nilai minimum 23 dan maksimum 100, dengan rata-rata nilai dari 482 anggota yaitu 65.8. Berikut visualisasi datanya:

```python
sns.histplot(data, x='reading score', hue='gender')
```

c. Nilai menulis

```python
data.groupby('gender')['writing score'].agg(['min', 'max', 'mean', 'count'])
```

Dilihat dari data berikut, perempuan memiliki nilai minimum 10 dan maksimum 100, dengan rata-rata nilai dari 518 anggota yaitu 72.5. Sedangkan laki-laki memiliki nilai minimum 15 dan maksimum 100, dengan rata-rata nilai dari 482 anggota yaitu 63.3. Berikut visualisasi datanya:

```python
sns.histplot(data, x='writing score', hue='gender')
```

Kesimpulan: laki-laki cenderung lebih pintar matematika dibandingkan perempuan, namun perempuan lebih mahir dalam membaca dan menulis
